```python
def split_on_number(s):
    parts = s.split('_')
    for i, p in enumerate(parts):
        if p and p[0].isdigit():
            return set(parts[:i]), '_'.join(parts[i:])
    return set(parts), ''
```

```python
def diagnostic_tables(df, true_col="label", pred_col="order_pred",
                     collapse_unknown_preds=True):
    """
    Returns three tables:
      - counts: raw confusion matrix (true rows × predicted columns)
      - recall_pct: row-normalised confusion matrix (each row sums to 100%)
      - metrics: per-class precision, recall, F1, support
    """
    # The set of "real" classes is determined by the true labels
    true_classes = sorted(df[true_col].unique())

    if collapse_unknown_preds:
        # Map any predicted class not in true_classes to "<other>"
        # Useful for BioCLIP's unconstrained predictions
        df = df.copy()
        df[pred_col] = df[pred_col].where(
            df[pred_col].isin(true_classes), other="<other>"
        )
        pred_classes = true_classes + ["<other>"]
    else:
        pred_classes = sorted(df[pred_col].unique())

    # 1. Raw counts
    cm = confusion_matrix(df[true_col], df[pred_col], labels=pred_classes)
    # confusion_matrix uses 'labels' for both axes; restrict rows to true classes
    counts = pd.DataFrame(cm, index=pred_classes, columns=pred_classes)
    counts = counts.loc[true_classes, :]  # rows = actual true classes only
    counts.index.name = "true"
    counts.columns.name = "predicted"

    # 2. Row-normalised (per-class recall view)
    recall_pct = counts.div(counts.sum(axis=1), axis=0) * 100
    recall_pct = recall_pct.round(1)

    # 3. Per-class metrics
    report = classification_report(
        df[true_col], df[pred_col],
        labels=true_classes,         # restrict to real classes for precision/recall
        zero_division=0,
        output_dict=True,
    )
    metrics = pd.DataFrame(report).T
    metrics = metrics.loc[true_classes + ["macro avg", "weighted avg"]]
    metrics["support"] = metrics["support"].astype(int)
    metrics = metrics.round(3)

    return counts, recall_pct, metrics

```

### unsharp

```python
unsharp_current_all_path = "/Users/krolik/Documents/python_/MScThesis/diopsis_apollo_all/md/preprocessing/order/last_attempt_full_run/bioclip_order_unsharp_current_preprocessed_all.csv"

unsharp_current_all = pd.read_csv(unsharp_current_all_path)
unsharp_current_all = unsharp_current_all.drop(columns=['kingdom','phylum','class'])
unsharp_current_all[['labels_set', 'image_id']] = unsharp_current_all['file_name'].apply(
    lambda s: pd.Series(split_on_number(s))
)

unsharp_current_all = unsharp_current_all.rename(columns={'order':'order_redone','score':'score_redone'}).drop(columns=['labels_set'])
unsharp_current_all = unsharp_current_all.merge(meta,on='image_id', how='left')
unsharp_current_all['rank_order_stimmt_redo'] = [0 if unsharp_current_all['order_redone'][i]==unsharp_current_all['label'][i] else 1 for i in range(unsharp_current_all.shape[0])]
print('all:', unsharp_current_all['rank_order_stimmt_redo'].value_counts())
```

```python
unsharp_per_order = unsharp_current_all.groupby("label").agg(
    n=("score", "size"),
    n_misclassified=("rank_order_stimmt", "sum"),
    n_rescued=("rank_order_stimmt_redo", 
                lambda x: ((unsharp_current_all.loc[x.index, "rank_order_stimmt"] == 1) & 
                            (x == 0)).sum()),
    n_damaged=("rank_order_stimmt_redo", 
                lambda x: ((unsharp_current_all.loc[x.index, "rank_order_stimmt"] == 0) & 
                            (x == 1)).sum()),
)
unsharp_per_order["net"] = unsharp_per_order["n_rescued"] - unsharp_per_order["n_damaged"]
unsharp_per_order["original_error_rate"] = unsharp_per_order["n_misclassified"] / unsharp_per_order["n"]
unsharp_per_order["new_error_rate"] = (unsharp_per_order["n_misclassified"] - unsharp_per_order["n_rescued"] 
                                + unsharp_per_order["n_damaged"]) / unsharp_per_order["n"]
print("Unsharp\n", unsharp_per_order.sort_values("label", ascending=True))
```

# Preprocessing details

```python
TECHNIQUE_SPECS = {
    "unsharp_mask": {
        "target_property": "edge_density",
        "direction":       "low",
        "rationale":       "Enhances local edge contrast → directly targets low edge_density images "
                           "(the strongest Model 2 protective predictor)",
    },
    "clahe": {
        "target_property": "dynamic_range",
        "direction":       "low",
        "rationale":       "Increases local contrast → targets low dynamic_range images",
    },
    "denoise": {
        "target_property": "blur_score",
        "direction":       "high",
        "rationale":       "Removes noise → targets high blur_score (high Laplacian variance "
                           "= noise indicator in Model 2)",
    },
    "bilateral": {
        "target_property": "shadow",
        "direction":       "high",
        "rationale":       "Edge-preserving smooth lifts shadows → targets shadow-flagged images",
    },
    "white_balance": {
        "target_property": "colorfulness",
        "direction":       "low",
        "rationale":       "Neutralises colour casts → targets low colourfulness images",
    },
    "auto_contrast": {
        "target_property": "overexposed_pct",
        "direction":       "high",
        "rationale":       "Histogram stretching → targets overexposed images",
    },
    "gamma": {
        "target_property": "brightness_mean",
        "direction":       "low",
        "rationale":       "Brightens midtones → targets dark images",
    },
}
```

```python
def apply_clahe(img):
    """For low brightness_std (low contrast)."""
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    return cv2.cvtColor(cv2.merge((clahe.apply(l), a, b)), cv2.COLOR_LAB2BGR)

def apply_white_balance(img):
    """Gray-world white balance, for color casts (twilight, low colorfulness)."""
    img_f = img.astype(np.float32)
    means = img_f.reshape(-1, 3).mean(axis=0)
    gray = means.mean()
    # Avoid division by zero on near-black channels
    means_safe = np.where(means < 1, 1, means)
    img_f = img_f * (gray / means_safe)
    return np.clip(img_f, 0, 255).astype(np.uint8)

def apply_auto_contrast(img):
    """For over/underexposed images. PIL autocontrast with 1% cutoff."""
    pil = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    return cv2.cvtColor(np.array(ImageOps.autocontrast(pil, cutoff=1)),
                        cv2.COLOR_RGB2BGR)

def apply_unsharp_mild(img):
    """Milder: smaller amplitude."""
    blurred = cv2.GaussianBlur(img, (0, 0), sigmaX=2)
    return cv2.addWeighted(img, 1.25, blurred, -0.25, 0)  # amount=0.25

def apply_unsharp_current(img):
    """Current: original setting."""
    blurred = cv2.GaussianBlur(img, (0, 0), sigmaX=2)
    return cv2.addWeighted(img, 1.5, blurred, -0.5, 0)  # amount=0.5

def apply_unsharp_aggressive(img):
    """More aggressive: larger amplitude, broader sigma."""
    blurred = cv2.GaussianBlur(img, (0, 0), sigmaX=3)
    return cv2.addWeighted(img, 2.0, blurred, -1.0, 0)  # amount=1.0

def apply_gamma(img, gamma=1.2):
    """For brightness imbalance. gamma > 1 brightens shadows."""
    inv = 1.0 / gamma
    table = np.array([((i / 255.0) ** inv) * 255 for i in range(256)]).astype(np.uint8)
    return cv2.LUT(img, table)

def apply_bilateral(img):
    """For shadow effects; edge-preserving smooth lifts midtones."""
    return cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)

def apply_denoise(img):
    """For noisy (night) images. Non-Local Means denoising."""
    return cv2.fastNlMeansDenoisingColored(img, None,
                                            h=10, hColor=10,
                                            templateWindowSize=7,
                                            searchWindowSize=21)
```
